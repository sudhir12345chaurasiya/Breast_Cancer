## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## loading the dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
print(cancer['DESCR'])
```

```python
print(cancer['target'])
```

```python
## malignant (encoded 0) and how many benign (encoded 1)
print(cancer['target_names'])
```

```python
print(cancer['feature_names'])
```

```python
cancer['data'].shape
```

## storing it as DataFrame

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns= np.append(cancer['feature_names'],['target']))
```

```python
df_cancer.head()
```

```python
# malignant (encoded 0) and how many benign (encoded 1)
sns.countplot(df_cancer['target'])
```

## maligant and benign count in our dataset

```python
malignant_count = len(df_cancer[df_cancer['target'] == 0])
benign_count = len(df_cancer[df_cancer['target'] == 1])
    
index = ['malignant', 'benign']
    
target = pd.Series(data=[malignant_count, benign_count], index=index)
    
```

```python
target
```

## Dividing our data into dependent(y) and independent(X) variables

#### dropping the target from the dataset and labeling rest as independent variable

```python
X = df_cancer.drop(['target'],axis =1)
```

```python
X
```

#### dependent variable

```python
y= df_cancer['target']
```

```python
y
```

## splitting into train and test datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

## Training our model on SVM(support vector classification) algorithm

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
svc_model= SVC()
```

```python
svc_model.fit(X_train,y_train)
```

```python
y_predict =svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
sns.heatmap(cm ,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

## Improving our model efficiency

### for train

```python
X_train
```

```python
min_train =X_train.min()
min_train
```

```python
range_train =(X_train - min_train).max()
range_train
```

### Scaling

```python
X_train_scaled =(X_train-min_train)/range_train
X_train_scaled
```

### for test

```python
min_test =X_test.min()
range_test =(X_test - min_test).max()
X_test_scaled =(X_test-min_test)/range_test
```

```python
min_test
```

```python
range_test
```

```python
X_test_scaled
```

```python
y_predict =svc_model.predict(X_test_scaled)
```
